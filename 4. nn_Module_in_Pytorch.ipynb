```python
import torch
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler,LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/gscdit/Breast-Cancer-Detection/refs/heads/master/data.csv')
df.head()
```

```python
df.drop(columns=['id','Unnamed: 32'],inplace=True)
df.head()
```

```python
x_train,x_test,y_train,y_test = train_test_split(df.iloc[:,1:],df.iloc[:,0],test_size=0.2)
```

```python
x_train.shape
```

```python
x_test.shape
```

## Scaling

```python
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
x_train
```

## Encoding

```python
encoder = LabelEncoder()
y_train = encoder.fit_transform(y_train)
y_test = encoder.transform(y_test)
```

```python
y_train
```

```python
type(x_train)
```

## Convert Numpy to Pytorch Tensors

```python
x_train_tensor = torch.from_numpy(x_train).float()
x_test_tensor = torch.from_numpy(x_test).float()
y_train_tensor = torch.from_numpy(y_train).float()
y_test_tensor = torch.from_numpy(y_test).float()
```

```python
type(x_train_tensor)
```

## Define Model

```python
import torch.nn as nn

class MySimpleNN(nn.Module):
  def __init__(self,num_features):
    super().__init__()
    self.network = nn.Sequential(
        nn.Linear(num_features,3),
        nn.ReLU(),
        nn.Linear(3,1),
        nn.Sigmoid(),
    )

  def forward(self,features):
    out = self.network(features)

    return out
```

```python
epochs = 50
learning_rate = 0.1
```

```python
loss_function = nn.BCELoss()
```

```python
# create model
model = MySimpleNN(x_train_tensor.shape[1])

# define optimizer
optimizer = torch.optim.SGD(model.parameters(),lr=learning_rate)
# model.parameters() => returns an iterator containing all learnable weights and biases that PyTorch computes gradients for and updates during training.

for epoch in range(epochs):
  # forward pass
  y_pred = model(x_train_tensor)

  # calculate loss
  loss = loss_function(y_pred,y_train_tensor.view(-1,1))
  # .view is same as .reshape in numpy
  # -1 tells the .view to decide the shape at that position on its own

  # clear gradients => so that gradients dont accumulate
  optimizer.zero_grad()

  # backward pass
  loss.backward()

  # parameters update
  optimizer.step()

  # print loss in each epoch
  print(f'Epoch: {epoch + 1}, Loss: {loss.item()}')
```

# Model Evaluation

```python
with torch.no_grad():
  y_pred = model(x_test_tensor)
  y_pred = (y_pred > 0.5).float()
  accuracy = (y_pred == y_test_tensor).float().mean()
  print(f'accuracy : {accuracy.item():.4f}')
```
